```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
```

```python
df_untied = pd.read_csv('./not_tied.csv')
df_untied.head()
```

```python
x = df_untied['cluster_num'].values.tolist()
y = df_untied['iterations'].values.tolist()
train_LL = df_untied['train_LL'].values.tolist()
dev_LL = df_untied['dev_LL'].values.tolist()
```

```python
%matplotlib inline
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.set_xlabel('cluster number')
ax.set_ylabel('iterations')
ax.set_zlabel('train log likelihood')
ax.scatter3D(x, y, train_LL, label='not tied graph')
```
